## Dataset parsing

```python
import json
from pathlib import Path
from pymatgen.core import Lattice, Structure, Composition

def _basis(crystal: str):
    ''' returns the basis coordinates, either bcc or fcc '''
    c = crystal.strip().lower()
    if c == "bcc":
        return [(0,0,0), (0.5,0.5,0.5)]
    elif c == "fcc":
        return [(0,0,0), (0,0.5,0.5), (0.5,0,0.5), (0.5,0.5,0)]
    raise ValueError(f"Unsupported crystal structure: {crystal}")

def _species_dict(comp_str: str): 
    ''' returns a dictionary of the species, normalized to 1.0 '''
    comp = Composition(comp_str)
    total = sum(comp.get_el_amt_dict().values())
    return {el: amt/total for el, amt in comp.get_el_amt_dict().items()}

def record_to_structure(rec: dict):
    a = float(rec["Lattice constant"])   # Å
    latt = Lattice.cubic(a)
    coords = _basis(rec["Crystal structure"])
    species = _species_dict(rec["Composition"])

    # make one disordered site per basis position
    struct = Structure(latt, [species]*len(coords), coords, validate_proximity=True)

    # stash other properties as annotations
    struct.annotations = {k: v for k, v in rec.items() if k not in {"Composition", "Crystal structure", "Lattice constant"}}

    return struct

def load_structures(json_path: str):
    data = json.loads(Path(json_path).read_text())
    structures = []
    for rec in data:
        try:
            structures.append(record_to_structure(rec))
        except Exception as e:
            # keep going but report which entry failed
            print(f"Skipped entry with Composition={rec.get('Composition')} due to: {e}")
    return structures

# Example:
structs = load_structures("data/dataset_Zhang_Compositiondesignhighentropy_2022.txt")
print(f'total length of list: {len(structs)}')
# structs[0].to(fmt="poscar")  # or Structure.to(filename="POSCAR")
```

```python
%%time
import matcalc as mtc
from sklearn.metrics import r2_score, mean_absolute_error
UNITS_MULTIPLIER = 160.2176621 # unit conversions for eV/A^3 to GPa

iaps = ['M3GNet-MatPES-PBE-v2025.1-PES', 'TensorNet-MatPES-PBE-v2025.1-PES']


shorter_list = mat_list
structs = [Structure.from_dict(entry['structure']) for entry in shorter_list]


for iap in iaps:
    calculator = mtc.ElasticityCalc(iap, relax_structure=True)
    elasticity_data = list(calculator.calc_many(structs, n_jobs=-1))
    bulk_pred = []
    bulk_actual = []
    shear_pred = []
    shear_actual = []

    for index, props in enumerate(elasticity_data):
        mat = shorter_list[index]
        bulk_pred.append(props['bulk_modulus_vrh'] * UNITS_MULTIPLIER)
        bulk_actual.append(mat['bulk_modulus_vrh'])
        shear_pred.append(props['shear_modulus_vrh'] * UNITS_MULTIPLIER)
        shear_actual.append(mat['shear_modulus_vrh'])

    results.append({
        'iap': iap,
        'bulk_actual': bulk_actual,
        'bulk_pred': bulk_pred,
        'shear_actual': shear_actual,
        'shear_pred': shear_pred
    })

for res in results:
    bulk_r2 = r2_score(res['bulk_actual'], res['bulk_pred'])
    shear_r2 = r2_score(res['shear_actual'], res['shear_pred'])
    print(res['iap'])
    print(f"Bulk Modulus R^2: {bulk_r2:.4f}")
    print(f"Shear Modulus R^2: {shear_r2:.4f}\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

# Flag to exclude outliers (set to True to exclude)
exclude_outliers = True

bulk_errors = []
shear_errors = []
model_names = []

# Reorder results: M3GNet, CHGNet, TensorNet


for res in results:
    bulk_actual = np.array(res['bulk_actual'])
    bulk_pred = np.array(res['bulk_pred'])
    shear_actual = np.array(res['shear_actual'])
    shear_pred = np.array(res['shear_pred'])

    bulk_pct_error = 100 * (bulk_pred - bulk_actual) / bulk_actual
    shear_pct_error = 100 * (shear_pred - shear_actual) / shear_actual

    bulk_errors.append(bulk_pct_error)
    shear_errors.append(shear_pct_error)
    model_names.append(res['iap'])

order = ['M3GNet','CHGNet', 'TensorNet']
ordered_bulk_errors = []
ordered_shear_errors = []
ordered_model_names = []

for model in order:
    for i, name in enumerate(model_names):
        if model.lower() in name.lower():
            ordered_bulk_errors.append(bulk_errors[i])
            ordered_shear_errors.append(shear_errors[i])
            ordered_model_names.append(name)

# Bulk modulus box plot (horizontal)
plt.figure(figsize=(8,5))
sns.boxplot(data=ordered_bulk_errors, orient="h", showfliers=not exclude_outliers)
plt.yticks(range(len(ordered_model_names)), ordered_model_names)
plt.xlabel('Bulk Modulus % Error')
plt.title('Bulk Modulus Percentage Error Boxplot')
plt.tight_layout()
plt.show()

# Shear modulus box plot (horizontal)
plt.figure(figsize=(8,5))
sns.boxplot(data=ordered_shear_errors, orient="h", showfliers=not exclude_outliers)
plt.yticks(range(len(ordered_model_names)), ordered_model_names)
plt.xlabel('Shear Modulus % Error')
plt.title('Shear Modulus Percentage Error Boxplot')
plt.tight_layout()
plt.show()
```
